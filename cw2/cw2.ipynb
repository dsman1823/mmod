Возспользуемся методом обратных функций. Начнем с нахождения функции, обратной функции распределени Парето. 

Искомая обратная функция равна $ x = F^{-1}(u) = \frac{x^{_{m}}}{(1 - u)^{1/k}}, \forall x > x_{m}  $, где $ u \sim Uni(0, 1) $
